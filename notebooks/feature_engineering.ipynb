```python
import pandas as pd
```

```python
data = pd.read_csv("../dataset/data_clean.csv")
```

```python
data['Age_Group'] = pd.cut(
    data['Age'],
    bins=[0,25,40,60],
    labels=['Young', 'Adult', 'Senior']
)
```

```python
data
```

```python
threshold = data['Purchase_Amount'].quantile(0.75)

data['Customer_Value'] = data['Purchase_Amount'].apply(
    lambda x: 'High Value' if x >= threshold else 'Low Value'
)
```

```python
data['Decision_Speed'] = data['Time_to_Decision'].apply(
    lambda x: 'Fast' if x <= 2 else 'Slow'
)
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
data.to_csv('../dataset/ecommerce_cleaned.csv', index=False)
```
